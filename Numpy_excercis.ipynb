```
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/realpython/python-data-cleaning/master/Datasets/BL-Flickr-Images-Book.csv')
df.head()
```

```
# Non-stardard missing values 
# -, --, na, n/a
# This will replace the non-standard missing values with standard missing value (NaN)
missing_values = ['-', '--', 'na', 'n/a']
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/property.csv', na_values = missing_values)
df.head()
```

```
df['NUM_BATH'] = pd.to_numeric(df['NUM_BATH'], errors='coerce')
df
```

```
df['ST_NAME'].fillna(125,inplace=True)
df
```

```
df.loc[4,'ST_NUM']=125
df
```

```
median=df['NUM_BEDROOMS'].median()
df['NUM_BEDROOMS'].fillna(median,inplace=True)
df
```
